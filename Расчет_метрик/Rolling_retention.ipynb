```python
import os
from datetime import datetime
from collections import defaultdict

def rolling_retention_30(file_entry, file_registration):
    if not os.path.exists(file_entry) or not os.path.exists(file_registration):
        print(f"❌ Файл(ы) не найден(ы): {file_entry}")
        return None

    registrations = {} # Словарь регистраций
    entries = defaultdict(list) # Словарь со значениями по умолчанию - заходов

    with open('registrations.csv', 'r', encoding='utf-8') as f:
        next(f)
        for line in f:
            if line.strip():
                user_id, date_str = line.strip().split(';')
                date = datetime.strptime(date_str, '%Y-%m-%d')

                if date.year == 2021 and date.month == 1:
                    registrations[user_id] = date

    with open('entries.csv', 'r', encoding='utf-8') as f:
        next(f)
        for line in f:
            if line.strip():
                user_id, entry_date_str = line.strip().split(';')
                entries[user_id].append(datetime.strptime(entry_date_str, '%Y-%m-%d')) # defaultdict - автоматически создает значение для нового ключа

    total = len(registrations)
    retained = 0 # Счетчик пользователей, вернувшихся 30 и более дня

    for user_id, reg_date in registrations.items():
        user_retained = False

        for entry_date in entries.get(user_id, []):
            days = (entry_date - reg_date).days

            if days >= 30:
                user_retained = True
                break

        if user_retained:
            retained+=1

    rolling_retention=(retained/total*100) if total> 0 else 0
    print(f"Метрика Rolling retention 30 дня - {round(rolling_retention, 5)}")
    return round(rolling_retention, 5)

rolling_retention_30('entries.csv', 'registrations.csv')

```
