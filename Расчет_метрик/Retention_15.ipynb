```python
import os
from datetime import datetime
from collections import defaultdict # Для создания словаря со значениями по умолчанию - списками

def retention_15(file_entry='entries.csv', file_registration='registrations.csv'):
    if not os.path.exists(file_entry) or not os.path.exists(file_registration):
        print(f"❌ Файл(ы) не найден(ы): {file_entry}")
        return None

    registrations = {} # Словарь регистраций
    entries = defaultdict(list) # Словарь со значениями по умолчанию - заходов

    # Чтение registrations.csv с разделителем ;
    with open(file_registration, 'r', encoding='utf-8') as f:
        next(f)  # Пропуск заголовка
        for line in f:
            if line.strip(): # Проверка, что строка непуста - удаление пробелов
                user_id, reg_date_str = line.strip().split(';') # Разделитель ;
                reg_date = datetime.strptime(reg_date_str, '%Y-%m-%d') # Преобразование строки в datetime
                if reg_date.year == 2021 and reg_date.month == 1:
                    registrations[user_id] = reg_date # Добавление записи в словарь registrations

    with open(file_entry, 'r', encoding='utf-8') as f:
        next(f)
        for line in f:
            if line.strip():
                user_id, entry_date_str = line.strip().split(';')
                entries[user_id].append(datetime.strptime(entry_date_str, '%Y-%m-%d')) # defaultdict - автоматически создает значение для нового ключа

    retention_window = 15
    total_registered = len(registrations)
    retained_count = 0  # Счетчик пользователей, вернувшихся до 15 дня

    for user_id, reg_date in registrations.items():
        for entry_date in entries.get(user_id, []):
            days_diff = (entry_date - reg_date).days
            if retention_window == days_diff:
                retained_count += 1
                break

    retention_15_day = (retained_count/total_registered * 100) if total_registered>0 else 0
    print(f"Метрика Retention 15 дня - {round(retention_15_day, 5)}")
    return round(retention_15_day, 5)



retention_15('entries.csv', 'registrations.csv')
```
