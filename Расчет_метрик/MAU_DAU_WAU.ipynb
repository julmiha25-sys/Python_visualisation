```python
import os
from datetime import datetime
from collections import defaultdict

def metric(file_entry, file_registration):
    if not os.path.exists(file_entry) or not os.path.exists(file_registration):
        print(f"❌ Файл(ы) не найден(ы): {file_entry}")
        return None

    registrations = {} # Словарь регистраций
    entries_month = defaultdict(list) # Словарь со значениями по умолчанию - заходов - за месяц
    entries_week = defaultdict(list) # Словарь со значениями по умолчанию - заходов - за неделю
    entries_day = defaultdict(list) # Словарь со значениями по умолчанию - заходов - за день

    with open('registrations.csv', 'r', encoding='utf-8') as f:
        next(f)
        for line in f:
            if line.strip():
                user_id, date_str = line.strip().split(';')
                date = datetime.strptime(date_str, '%Y-%m-%d')

                registrations[user_id] = date

    with open('entries.csv', 'r', encoding='utf-8') as f:
        next(f)
        for line in f:
            if line.strip():
                user_id, entry_date_str = line.strip().split(';')
                entry_date = datetime.strptime(entry_date_str, '%Y-%m-%d')
                if entry_date.year == 2021 and entry_date.month == 12:
                  entries_month[user_id].append(datetime.strptime(entry_date_str, '%Y-%m-%d')) # defaultdict - автоматически создает значение для нового ключа
                  if entry_date.day>=25 and entry_date.day <=31:
                    entries_week[user_id].append(datetime.strptime(entry_date_str, '%Y-%m-%d'))
                    if entry_date.day==31:
                      entries_day[user_id].append(datetime.strptime(entry_date_str, '%Y-%m-%d'))

    dec_mau = len(entries_month)
    dec_wau = len(entries_week)
    dec_dau = len(entries_day)
    print(f" Метрики: \n MAU - {dec_mau},\n WAU - {dec_wau},\n DAU - {dec_dau} ")
    return dec_mau, dec_wau, dec_dau

metric('entries.csv', 'registrations.csv')
```
