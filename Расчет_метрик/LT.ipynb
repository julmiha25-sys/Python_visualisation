```python
import os
from datetime import datetime
from collections import defaultdict

def lifetime(file_entry, file_registration, max_days=365):

    if not os.path.exists(file_entry) or not os.path.exists(file_registration):
        print(f"❌ Файл(ы) не найден(ы)")
        return None

    registrations = {} # Словарь регистраций
    entries = defaultdict(list) # Словарь со значениями по умолчанию - список

    with open(file_registration, 'r', encoding='utf-8') as f:
        next(f)
        for line in f:
            if line.strip():
                user_id, date_str = line.strip().split(';')
                date = datetime.strptime(date_str, '%Y-%m-%d')
                registrations[user_id] = date

    with open(file_entry, 'r', encoding='utf-8') as f:
        next(f)
        for line in f:
            if line.strip():
                user_id, entry_date_str = line.strip().split(';')
                entries[user_id].append(datetime.strptime(entry_date_str, '%Y-%m-%d')) # defaultdict - автоматически создает значение для нового ключа

    total_users = len(registrations)
    if total_users == 0:
        return 0

    retained_by_day = defaultdict(int) # Словарь со значениями по умолчанию - int (для подсчета)

    for user_id, reg_date in registrations.items():
        unique_days = set() # Уникальные дни активности для пользователя
        for entry_date in entries.get(user_id, []):
            days_diff = (entry_date - reg_date).days
            if 0 <= days_diff <= max_days:
                unique_days.add(days_diff)  # {0,1,3,5,7,...}

        for day in unique_days:
            retained_by_day[day] += 1

    lifetime = sum(count/total_users for count in retained_by_day.values())
    print(f"Метрика LifeTime - {round(lifetime, 4)}")
    return round(lifetime, 4)

lifetime('entries.csv', 'registrations.csv')
```
