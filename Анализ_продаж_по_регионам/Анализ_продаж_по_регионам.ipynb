```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
df1=pd.read_csv('products_data.csv')
df2=pd.read_csv('sales_data1.csv')
df=df2.merge(df1,left_on='product_id',right_on='product_id',how='left')
df3=df.groupby(['product_id','manufacturer','region']).agg({'quantity':'sum'}).reset_index()
g=sns.pairplot(df3, hue='region')
numeric_cols=df3.select_dtypes(include=['number']).columns.tolist()
quantity_idx=numeric_cols.index('quantity') if 'quantity' in numeric_cols else -1
if quantity_idx!=-1:
  g.axes[quantity_idx, quantity_idx].set_xlim(left=0)
  for i in range(len(numeric_cols)):
    g.axes[i, quantity_idx].set_xlim(left=0)
plt.show()
```
