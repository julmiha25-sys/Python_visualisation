```python
import os
import re

def auto_renewal_sub(file_path):
    if not os.path.exists(file_path):
        print(f"❌ Файл не найден: {file_path}")
        return None

    with open(file_path, 'r', encoding='utf-8') as f:
        lines=f.readlines()

    date_num_dict={} # Cловарь {дата: число} c последним числом пользователей с автопродлением подписки
    for line in lines:
        if 'количество людей с автопродлением подписки:' in line:
            date_match=re.search(r'\d{4}-\d{2}-\d{2}', line) # Поиск даты
            if date_match: # Объект Match
                date=date_match.group(0) # Берём первую найденную дату как строку
                try:
                    number=int(line.split(':')[-1].strip()) # Ищем посдений элемент - число
                    date_num_dict[date]=number  # Перезаписываем словарь
                except ValueError:
                    continue

    final_numbers=[date_num_dict[date] for date in sorted(date_num_dict)] # Сортируем по дате и извлекаем значения
    cum_sum = [sum(final_numbers[:i+1]) for i in range(len(final_numbers))] # Суммируем кол-во пользователей
    cum_count = [i+1 for i in range(len(final_numbers))] # Суммируем кол-во дней
    cum_mean = [round(cum_sum[i]/cum_count[i],2) for i in range(len(final_numbers))] # Накопительное среднее

    cum_med = [] # Список с накопительной медианой
    for i in range(len(final_numbers)):
        slice_data = sorted(final_numbers[:i+1])
        mid=i//2
        if i%2==0:  # Четное количество элементов
            median=int(slice_data[mid])
        else:  # Нечетное количество элементов
            median = int(((slice_data[mid]+slice_data[mid+1])/2))
        cum_med.append(median)
    filename = 'auto_renewal_sub.txt' # Имя выходного файла
    print(f"\n Сохранение результатов в файл {filename}...")
    with open(filename, 'w', encoding='utf-8') as f:
       f.write(f"Среднее: {cum_mean}\n")
       f.write(f"Медиана: {cum_med}\n")
       print(f"Файл сохранен: {filename}")
    return


auto_renewal_sub('auto_purchase.log')
```
